# Private attributes

```python
class Patient:
    def __init__(self, name, diagnosis) -> None:
        self._name = name
        self.__diagnosis = diagnosis

    def __repr__(self) -> str:
        return f"Patient({self._name}, {self.__diagnosis}) "

    
        
p1=Patient("Gore Bord", "Migraine")

print(p1)

# it works bit should not do this
p1._name = "Gree Board"
print(p1)

# try to change __diagnosis
p1.__diagnosis = "cold"
print(p1)

print(p1.__dict__)
# Python does name mangling ->> change __diagnosis to _Patient__diagnosis

p1._Patient__diagnosis = "Fever"
print(p1)
```

# Matplotlib
- matlab - style
- OOP approach

```python
# Setup
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5,5)
f = np.sin(x)
print(x)
print(f)

```

```python
plt.figure(dpi=100)
plt.plot(x,f)
plt.title("sin(x)")
plt.xlabel("x")
plt.ylabel("y")
```

```python
plt.figure(dpi = 100, figsize = (10,3))
plt. subplot(1,2,1) # (rows, cols, panel number)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))


plt.subplot(1,2,2)
plt.plot(x, np.cos(x))
# plt.show


```

# OOP approach

```python
fig = plt.figure(dpi=100)
# ax object can have a plot , axis, labels, title.....
ax = plt.axes()
print(fig)

ax.plot(x,f)
ax.set(xlabel = "x", ylabel="y", title="$\sin{x}$", xlim=(0.5))
```

```python
fig, ax = plt.subplots(1,2,dpi=100, figsize=(10,4))
rng = np.random.RandomState(43)
data_normal_dist = rng.randn(1000)

ax[0].hist(data_normal_dist, bins=30, alpha=0.5)
ax[0].set(title="Histogram of normal distributed data", xlabel="x", ylabel="y")

bubble = ax[1].scatter(x=rng.randn(100), y=rng.randn(100), c=rng.randn(100), alpha=0.7, cmap= "cividis", s=1000*rng.rand(100))
ax[1].set(title="Bubble chart", xlabel="x", ylabel="y")
fig.colorbar(bubble)

plt.savefig("Beautiful_plots.png", facecolor="white")
```
