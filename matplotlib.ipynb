# Private attributes

```python
class Patient:
    def __init__(self, name, diagnosis) -> None:
        self._name = name
        self.__diagnosis = diagnosis

    def __repr__(self) -> str:
        return f"Patient({self._name}, {self.__diagnosis}) "

    
        
p1=Patient("Gore Bord", "Migraine")

print(p1)

# it works bit should not do this
p1._name = "Gree Board"
print(p1)

# try to change __diagnosis
p1.__diagnosis = "cold"
print(p1)

print(p1.__dict__)
# Python does name mangling ->> change __diagnosis to _Patient__diagnosis

p1._Patient__diagnosis = "Fever"
print(p1)
```

```python
# Setup
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5,5)
f = np.sin(x)
print(x)
print(f)

```

```python
plt.figure(dpi=100)
plt.plot(x,f)
plt.title("sin(x")
plt.xlabel("x")
plt.ylabel("y")
```

```python
plt.figure(dpi = 100, figsize = (10,3))
plt. subplot(1,2,1) # (rows, cols, panel number)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))


plt.subplot(1,2,2)
plt.plot(x, np.cos(x))
# plt.show


```
